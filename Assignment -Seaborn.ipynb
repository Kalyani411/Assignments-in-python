```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
taxis = sns.load_dataset('taxis')
taxis.head()
```

```python
taxis["fare"].unique()
```

```python
sns.distplot(taxis['fare'], bins=6)
```

```python
sns.distplot(taxis['fare'], bins=range(3,10,1))
plt.xticks(range(3,9,1))
```

```python
sns.distplot(taxis["fare"], kde=False, bins=range(3,10,1))
```

```python
sns.distplot(taxis["fare"], hist=False, bins=range(3,10,1))
plt.xticks(range(3,9,1))
```

```python
sns.countplot(x = 'fare', data=taxis)
```

```python
sns.countplot(x = 'fare', data=taxis, color='red')
```

```python
taxis['dropoff_zone'].unique()
```

```python
sns.countplot(x = 'fare', data=taxis, color='red', hue='color')
```

```python
sns.countplot(x = 'fare', data=taxis, color='blue', hue='color')
```

```python
sns.countplot(x = 'fare', data=taxis, color='blue', hue='color')
plt.legend(loc = 'lower right')
```

```python
sns.boxplot(y='tolls', data=taxis)
plt.xlabel('Tolls')
```

```python
sns.boxplot(y='tolls',x='color', data=taxis)
```

```python
sns.boxplot(y='tolls',x='fare', data=taxis)
```

```python
# Scatter plot
sns.jointplot(x='distance',y='tolls', data=taxis)
```

```python
sns.scatterplot(x='distance',y='tolls', data=taxis)
```

```python
sns.scatterplot(y='tolls',x='distance', data=taxis, alpha=0.2)
```

```python
sns.scatterplot(y='tolls',x='distance', hue='color', data=taxis)
```

```python
plt.subplot(2,2,1)
sns.scatterplot(y='tolls',x='distance', hue='color', data=taxis)

plt.subplot(2,2,2)
sns.scatterplot(y='tolls',x='distance', data=taxis, alpha=0.2)

plt.subplot(2,2,3)
sns.boxplot(y='tolls',x='color', data=taxis)
```
